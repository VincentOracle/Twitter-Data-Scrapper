# Kaggle 2023  Machine Learning Challenge

This machine learning data challenge involved a regression task. The target variable is 'spend' in dollars by a customer at a shopping mall. There are 131 independent variables that may have an impact on the target variable 'spend'. First variable 'Id' identifies each customer uniquely in the shopping mall.

The challenge is to correctly predict value for 'spend' based on independent variables.

```python
#Import the necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python

# Load the data
train_df = pd.read_csv('C:\\Users\\n\\Downloads\\train.csv\\train.csv')
test_df = pd.read_csv('C:\\Users\\n\\Downloads\\test.csv\\test.csv')
```

```python

# Prepare the data
X_train = train_df.drop(['Id', 'X109','spend','X51'], axis=1)
y_train = train_df['spend']
X_test = test_df.drop(['Id','X51','X109'], axis=1)
print(X_train )#View the prepared train dataset
```

```python
print(X_test)#View the prepared test dataset
```

#             Fit the decision tree regression model

```python
# Fit the decision tree regression model
model = DecisionTreeRegressor(random_state=42)
model.fit(X_train, y_train)
# Predict on the test set
y_pred = model.predict(X_test)
print(y_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
# Visualize the decision tree
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=X_train.columns, filled=True, fontsize=10)
plt.show()
```

```python

# Create a submission file
mysubmission_df = pd.DataFrame({'Id': test_df['Id'], 'Expected': y_pred})
print(mysubmission_df)
# submission_df.to_csv('submission.csv', index=False)
mysubmission_df.to_csv('C:/Users/n/Desktop/submission.csv', index=False)
```

# Evaluate the model on the training set

```python

# Evaluate the model on the training set
y_train_pred = model.predict(X_train)
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = mse_train ** 0.5
print(f'Training set RMSE: {rmse_train:.2f}')
```
